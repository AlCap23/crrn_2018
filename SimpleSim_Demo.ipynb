```python
import simple_sim.simulation_objects as Objects
import simple_sim.simulation_server as World
import pybullet_data
import os
import simple_sim.simulation_robot as Robots
```

```python
# Start a simulation server
server = World.simulationServer()
```

```python
# Add a heightmap
heightmap = Objects.SimHeightmap(server, './models/heightmap.obj')
```

```python
# Add the pybullet data folder to get more URDFs
data_path = pybullet_data.getDataPath()
# Get the R2D2 model
r2path = os.path.join(data_path, 'r2d2.urdf')
# Check if the model exists
os.path.exists(r2path)
```

```python
# Add the model
robot = Robots.Robot(server, r2path, basePosition = [0, 0, 2])
```

```python
# Set the gravity
server.setGravity([0, 0, -9.81])
```

```python
# Start the simulation
server.startSimulation()
```

```python
# Robot joint information is given via
robot.joints
# Robot link information is given via
robot.links
```
